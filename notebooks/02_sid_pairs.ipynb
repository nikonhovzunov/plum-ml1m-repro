```python
import pandas as pd
from pathlib import Path
```

```python
path = '../data/processed/splits/train.parquet'
```

```python
train = pd.read_parquet(path)
```

```python
train.head()
```

```python
train.shape
```

```python
train.tail(10)
```

```python
train['item_next'] = train['item_idx'].shift(-1)
train['user_next'] = train['user_id'].shift(-1)
```

```python
train.head()
```

```python
train.tail()
```

```python
pairs = train[(train['user_id'] == train['user_next']) & (train['item_next'].notna())]
```

```python
pairs.head()
```

```python
pairs[['item_next', 'user_next']] = pairs[['item_next', 'user_next']].astype(int)
```

```python
pairs
```

```python
co_pairs = pairs[['item_idx', 'item_next']]
```

```python
co_pairs.head()
```

```python
co_pairs.shape
```

```python
OUT_DIR = Path("../data/processed/sid_pairs")
OUT_DIR.mkdir(parents=False, exist_ok=True)
```

```python
out_path = OUT_DIR / "co_pairs_neighbors.parquet"
```

```python
co_pairs.to_parquet(out_path, index=False)
```

# CPT pipeline quickstart

Clean entrypoint for preparing PLUM-style CPT artifacts from a chosen SID matrix and optionally training GPT-2 on the resulting behavior/metadata mixture.

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if not (ROOT / "data").exists():
    ROOT = ROOT.parents[1]
sys.path.insert(0, str(ROOT))

import numpy as np

from src.cpt import (
    CPTArtifactPaths,
    CPTMixtureConfig,
    CPTPipeline,
    GPT2CPTTrainingConfig,
    train_gpt2_cpt,
)

ROOT
```

## Prepare CPT artifacts

Swap `SID_PATH` whenever you want to run CPT on a new SID version.

```python
SID_PATH = ROOT / "data" / "processed" / "SIDs_V1.npy"
RUN_NAME = "CPT_user_behavior_V2"

sids = np.load(SID_PATH)

pipeline = CPTPipeline(
    paths=CPTArtifactPaths(
        processed_dir=ROOT / "data" / "processed",
        raw_ml1m_dir=ROOT / "data" / "raw" / "ml-1m",
        artifacts_dir=ROOT / "data" / "processed" / "artifacts",
    )
)

prepared = pipeline.prepare_artifacts(
    sids=sids,
    run_name=RUN_NAME,
    base_tokenizer="gpt2",
    save=True,
)

prepared["stats"]
```
